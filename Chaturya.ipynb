## Step4: Cleaning the dataset

```python
# Drop columns with more than 50% missing data
columns_to_drop = missing_data_info[missing_data_info['Percentage of Total'] > 50].index
data.drop(columns=columns_to_drop, inplace=True)
```

```python
# Standardize column names by stripping leading and trailing spaces
data.columns = data.columns.str.strip()
```

```python
# Fill missing values in relevant columns
data['Populations Served'] = data['Populations Served'].astype('object').fillna('Unknown')
data['Program City'] = data['Program City'].astype('object').fillna('Unknown')
data['Program Zip'] = data['Program Zip'].astype('object').fillna('00000')
data['Program Address 1'] = data['Program Address 1'].fillna('Address Not Provided')
data['Program Phone'] = data['Program Phone'].fillna('Phone Not Provided')
```

```python
# Drop rows where essential information in 'Georeference' is missing
data.dropna(subset=['Georeference'], inplace=True)
```

```python
# Check for missing values in each column and their proportions
missing_values = data.isnull().sum().sort_values(ascending=False)
missing_percentage = (missing_values / len(data)) * 100

# Create a DataFrame to display missing values and percentages
missing_data_info = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage of Total': missing_percentage
})

print(missing_data_info)
```

```python
# Check for duplicates
data.duplicated().sum()
```

```python
# Convert 'Row Created Date Time' to datetime format
data['Row Created Date Time'] = pd.to_datetime(data['Row Created Date Time'], errors='coerce')
```

```python
# Standardize phone numbers by removing non-numeric characters and formatting as (XXX) XXX-XXXX
data['Agency Phone'] = data['Agency Phone'].str.replace(r'\D', '', regex=True).str.replace(r'(\d{3})(\d{3})(\d+)', r'(\1) \2-\3', regex=True)
data['Program Phone'] = data['Program Phone'].str.replace(r'\D', '', regex=True).str.replace(r'(\d{3})(\d{3})(\d+)', r'(\1) \2-\3', regex=True)
```

```python
# Validate ZIP codes and standardize format
data['Program Zip'] = data['Program Zip'].str.replace(r'\D', '', regex=True)  # Remove non-numeric characters
data['Program Zip'] = data['Program Zip'].str.pad(width=5, fillchar='0')  # Ensure minimum length of 5 digits
```

```python
# Check that 'Georeference' is in the format 'POINT (longitude latitude)'
valid_georeference = data['Georeference'].str.match(r'^POINT \(-?\d+(\.\d+)? -?\d+(\.\d+)?\)$')
data = data[valid_georeference]  # Keep only valid entries
```

```python
# Standardize text columns to title case
text_columns = ['Sponsor Name', 'Agency Name', 'Facility Name', 'Program Name', 'Program City', 'Program County', 'Program Region', 'Program Type Description', 'Program Category Description', 'Program Subcategory Description']
data[text_columns] = data[text_columns].apply(lambda x: x.str.title())
```

### Observation

Standardized Columns: Stripped spaces and applied title case for consistency
                                                            
Dropped Columns: Removed those with >50% missing data 
                                                            
Filled Missing Data: Used placeholders for non-critical columns; dropped rows missing Georeference
                                      
Date & Phone Formatting: Converted dates and standardized phone numbers

ZIP Codes & Georeference: Validated and standardized for accuracy

Duplicates: Checked and ensured data uniqueness

Text Standardization: Columns containing text were standardized to the title case for uniformity and improved readability

### Program Type vs. Population Served: Do treatment-based mental health programs serve more adults, while support services serve more children and adolescents?


```python
# Ensure all entries in 'Populations Served' are string type
data['Populations Served'] = data['Populations Served'].astype(str)

# Creating a new column to categorize the populations served
data['Population Category'] = data['Populations Served'].apply(lambda x: 'Adults Only' if 'Adults' in x and 'Children' not in x else 'Children/Adolescents Included')

# Re-grouping data for updated plots
program_population_counts = data.groupby(['Program Type Description', 'Population Category']).size().reset_index(name='Count')

# Plotting separate bar charts for each population category
plt.figure(figsize=(18, 16))

# Plot for 'Adults Only'
plt.subplot(2, 1, 1)
adult_programs = program_population_counts[program_population_counts['Population Category'] == 'Adults Only']
sns.barplot(data=adult_programs, x='Program Type Description', y='Count', hue='Population Category', palette=['#1f77b4']) 
plt.xticks(rotation=90)
plt.title('Program Types Serving Adults Only')
plt.xlabel('Program Type')
plt.ylabel('Number of Programs')
plt.legend(title='Population Category')

# Plot for 'Children/Adolescents'
plt.subplot(2, 1, 2)
child_programs = program_population_counts[program_population_counts['Population Category'] == 'Children/Adolescents Included']
sns.barplot(data=child_programs, x='Program Type Description', y='Count', hue='Population Category', palette=['#ff7f0e']) 
plt.xticks(rotation=90)
plt.title('Program Types Serving Children/Adolescents')
plt.xlabel('Program Type')
plt.ylabel('Number of Programs')
plt.legend(title='Population Category')

plt.tight_layout()
plt.show()
```

##### Programs Serving Adults Only:

The majority of treatment and specialized mental health programs predominantly serve adults. Examples include crisis intervention programs and adult-focused behavioral health services.
This indicates that adult-centric programs, particularly those requiring intensive treatment or specific interventions, are more numerous and tailored to adult needs

##### Programs Including Children/Adolescents:

Programs that serve children and adolescents alongside adults are generally less numerous compared to those serving only adults. These include support programs and community-based mental health services.
The presence of fewer child-focused or mixed-population programs suggests that while there are services available for younger populations, they may not be as specialized or extensive as those for adults

The data supports the hypothesis that treatment-based programs are primarily oriented toward adults, while broader support programs may serve a mixed population, including children and adolescents

### Program Type and Geographic Region: Are certain types of mental health programs (e.g., residential treatment) concentrated in certain areas/regions (e.g., New York City vs. Hudson River)?


```python
# Group data by 'Program Type Description' and 'Program Region' to see the distribution
program_region_distribution = data.groupby(['Program Type Description', 'Program Region']).size().reset_index(name='Program Count')

# Plotting the distribution using a bar plot
plt.figure(figsize=(22, 12))  # Increase figure size for better spacing
sns.barplot(data=program_region_distribution, x='Program Region', y='Program Count', hue='Program Type Description')
plt.title('Distribution of Program Types by Geographic Region')
plt.xlabel('Geographic Region')
plt.ylabel('Number of Programs')
plt.xticks(rotation=45)
plt.legend(title='Program Type Description', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust the layout with more control over spacing
plt.subplots_adjust(hspace=0.3, top=0.92, bottom=0.08)  # Adjust vertical and margin spacing

plt.show()
```

#### Let's improve the visualization by plotting top program types and regions separately for clarity


```python
# Filter data for the top 5 most common program types to make the plot clearer
top_program_types = program_region_distribution['Program Type Description'].value_counts().head(5).index
filtered_data = program_region_distribution[program_region_distribution['Program Type Description'].isin(top_program_types)]
```

```python
# Create a pivot table to transform the data for a stacked bar chart
pivot_data = filtered_data.pivot(index='Program Region', columns='Program Type Description', values='Program Count').fillna(0)

# Define a custom color palette
custom_colors =  ['#004d40', '#00796b', '#009688', '#26a69a', '#80cbc4'] 

# Plotting a stacked bar chart with the custom color palette
pivot_data.plot(kind='bar', stacked=True, figsize=(16, 8), color=custom_colors)
plt.title('Stacked Distribution of Top 5 Program Types by Geographic Region')
plt.xlabel('Geographic Region')
plt.ylabel('Number of Programs')
plt.xticks(rotation=45)
plt.legend(title='Program Type Description', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

##### Observations 

The Hudson River and New York City regions have a broad distribution of different program types, with significant numbers of programs such as Outreach and Specialty Mental Health Care Management.
Comprehensive PROS with Clinical Treatment programs are more evenly distributed across these regions but show a particularly higher presence in New York City.
    
##### Dominant Program Types by Region:

Specialty Mental Health Care Management has the highest concentration in New York City, suggesting that this area may have a more extensive network of specialized mental health services.
Advocacy/Support Services programs are notably prevalent in the Hudson River region, highlighting a focus on community and support-based mental health initiatives in this area.

##### Regional Specialization:

Central New York has a significant share of Outreach and Advocacy/Support Services, indicating regional specialization in accessible and community-centered programs.
Long Island shows a lower total number of programs but includes diverse types, suggesting a balance in available services.

##### Comparative Analysis:

Hudson River stands out for having a diverse set of high-count program types, making it a hub for various mental health services.
New York City also has diverse and numerous programs but with a particular emphasis on Specialty Mental Health Care Management.

                                                                                         
Certain mental health program types, such as Specialty Mental Health Care Management and Advocacy/Support Services, are concentrated in specific regions, especially urban centers like New York City and more densely populated areas like the Hudson River. This distribution reflects regional needs, population density, and resource allocation, providing insights into where services are more heavily focused.
